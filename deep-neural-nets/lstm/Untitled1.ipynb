```python
import numpy as np
#from matplotlib import cm
from matplotlib import pyplot as plt
from scipy import fftpack 
from sklearn.preprocessing import minmax_scale
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def pi_digits():
    memo = []
    filenm = "pi.txt"
    f = open(filenm, "r")
    for line in f.readlines():
        for ch in line:
            if ch != '\n':
                memo.append(int(ch))
    return np.array(memo)
```

```python
t = np.arange(0, 10000)
e = pi_digits()
```

```python
e
```

```python
plt.figure(num=None, figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')
plt.scatter(t[:150],e[:150])
```

```python
#e_ = e/10
t_ = t % 10
plt.figure(num=None, figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')
r = np.sqrt(t_**2 + e**2)
theta = np.arctan2(e,t_)
```

```python
plt.figure(num=None, figsize=(15, 6), dpi=80, facecolor='w', edgecolor='k')
plt.polar(theta[:300],r[:300])
```
